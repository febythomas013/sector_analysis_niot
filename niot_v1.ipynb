```python
correlation_matrix = (2 * (cosine_sim_matrix - cosine_sim_matrix.min()) /
                      (cosine_sim_matrix.max() - cosine_sim_matrix.min())) - 1
```

### Covariance Matrix Creation using Cosine Similarity Matrix

```python
# Convert cosine similarity to a covariance-like matrix
covariance_matrix = (cosine_sim_matrix - cosine_sim_matrix.mean()) / cosine_sim_matrix.std()

```

```python
# Print the covariance-like matrix
print("Covariance-Like Matrix:")
print(pd.DataFrame(covariance_matrix, index=df_transposed.index, columns=df_transposed.index))
```

### Negative and Positive Industry Relations using Eucledian Distance and Inverse Eucledian Distance (for Similarity Detection)

```python
# Convert the numpy array to a DataFrame with appropriate indices and columns
euclidean_distance_df = pd.DataFrame(euclidean_distance_df, index=df_transposed.index, columns=df_transposed.index)

# Set thresholds for negatively correlated and highly correlated sectors
negative_corr_threshold = -0.5  # Adjust this value as needed
high_corr_threshold = 0.7  # Adjust this value as needed
```

```python
# Set threshold for considering similar sectors
similar_threshold = 0.5  

# Initialize a list to store the results
results = []

# Identify similar sectors based on Euclidean distance
for sector1 in euclidean_distance_df.columns:
    for sector2 in euclidean_distance_df.columns:
        if sector1 != sector2:
            distance = euclidean_distance_df.loc[sector1, sector2]

            # Check for similarity based on the threshold
            if distance <= similar_threshold:
                similarity_score = 1 / (1 + distance)  # Inverse of Euclidean distance as a similarity score
                results.append((sector1, sector2, "Similar", similarity_score))
```

```python
# Create a DataFrame from the results
results_df = pd.DataFrame(results, columns=["Industry", "Related Industry", "Relation Type", "Similarity Score"])
```
